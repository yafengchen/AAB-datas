```python
import cv2
import glob
import os
```

```python
inputFolder = 'recipes'
os.mkdir('resized')
for img in glob.glob(inputFolder + "/*.png"):
    try:
        image = cv2.imread(img)
        imgResized = cv2.resize(image, (64, 64))
        cv2.imwrite("Resized" + img[7:], imgResized)
    except:
        print(img)
```

```python
image = cv2.imread('resized/CMPELQqh_nh.png')
```

```python
cv2.imshow('image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
